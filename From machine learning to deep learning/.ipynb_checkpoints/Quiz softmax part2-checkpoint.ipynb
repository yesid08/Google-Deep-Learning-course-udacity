```python
import numpy as np
def softmax(x):
    value = 1.*np.exp(x)/np.sum(np.exp(x), axis = 0)
    return value 
```

```python
import matplotlib.pyplot as plt
scores = np.array( [1.0, 2.0, 3.0] )
scores = scores * 10
print scores
print softmax(scores)
plt.plot(scores, softmax(scores), linewidth=2 )
plt.xlabel("Scores")
plt.ylabel("Probability of scores")
plt.show()
```

```python
# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])
scores = scores * 10
print "probabilidad de que ocurra alguna de las clases de scores"
print softmax(scores)
print "sumatoria de las columnas de cada clase del nparray scores"
print np.sum(softmax(scores), axis=0)
plt.plot(x,softmax(scores).T, linewidth=2 )
plt.legend(("X","1's like X","0,2's like x"))
plt.xlabel("Value of X")
plt.ylabel("softmax")
plt.show()
```
